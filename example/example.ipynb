# Parallel Coordinates Plot

The following is a simple example, how to create a parallel coordinates plot with the `pcp()` function.

```python
import sys
sys.path.insert(0,"..")

import matplotlib.pyplot as plt
from plot import load_csv, pcp
```

## Setup data

Required is a nested list `data` for the curves. Each curve is described by a list containing the data points on the y-axes. The labels for the y-axes is a list of strings `labels`.

```python
data, labels = load_csv("data.csv")
```

## Setup y-axes

Axes can be linear, logarithmic or categorical. Categorical axes are automatically recognized for string values. Often only logarithmic axes need to be specified. 

The axis limits are determined automatically from the data. If there is not enough data available or if you want to specify your own limits, these must be specified. 

```python
ytype = ["categorial", [], "log", "categorial", [], [], []]
ylim = [[], [], [1e-06, 0.001], [], [0, 1], [0, 0.25], [0.0, 1.0]]
```

## Plot and Save

Default plot without adjustments:

```python
fig = pcp(data, labels, alpha=0.8)
plt.show()
```

With custom `type`, `ylabels`, `yticks` and `ylim`:

```python
fig = pcp(data, labels, ytype, ylim, alpha=0.8)
plt.savefig("test.svg", format="svg", facecolor="white", bbox_inches="tight")
plt.show()
```
