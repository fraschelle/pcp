# Parallel Coordinates Plot

The following is a simple example, how to create a parallel coordinates plot with the `pcp()` function.

```python
import matplotlib.pyplot as plt
from plot import load_csv, pcp
```

## Setup data

Required is a nested list `data` for the curves. Each curve is described by a list containing the data points on the y-axes. The labels for the y-axes is a list of strings `labels`.

Load your data from a csv-file:

```python
data, labels = load_csv("data.csv")
```

Or define the data in here:

```python
data = [["201", 'sgd+momentum', 4e-06, 5, 0.24, 0.19, 0.182],
        ["121", 'rmsprop', 0.00039, 0, 0.43, 0.17, 0.244],
        ["201", 'sgd+nesterov', 2e-06, 3, 0.17, 0.08, 0.273],
        ["161", 'radam', 0.000532, 0, 0.96, 0.01, 0.279],
        ["121", 'sgd+momentum', 3e-06, 3, 0.63, 0.12, 0.296],
        ["121", 'sgd+nesterov', 2e-06, 2, 0.38, 0.16, 0.296],
        ["169", 'rmsprop', 0.000319, 0, 0.31, 0.05, 0.315],
        ["161", 'rmsprop', 0.000342, 0, 0.2, 0.03, 0.342],
        ["169", 'sgd+momentum', 2.9e-05, 5, 0.97, 0.04, 0.547],
        ["161", 'amsgrad', 2.6e-05, 1, 0.99, 0.23, 0.573],
        ["169", 'amsgrad', 3.3e-05, 4, 0.41, 0.19, 0.719],
        ["121", 'sgd+nesterov', 0.00051, 3, 0.52, 0.02, 0.621],
        ["201", 'radam', 3e-06, 2, 0.45, 0.04, 0.823],
        ["169", 'rmsprop', 1e-06, 4, 0.66, 0.11, 0.823],
        ["161", 'rmsprop', 7.2e-05, 3, 0.0, 0.02, 0.824],
        ["169", 'radam', 2e-06, 3, 0.48, 0.01, 0.825],
        ["161", 'radam', 3e-06, 3, 0.39, 0.04, 0.825],
        ["161", 'radam', 1e-06, 4, 0.67, 0.06, 0.866],
        ["161", 'sgd+momentum', 0.000133, 5, 0.17, 0.07, 0.826],
        ["201", 'adam', 1.2e-05, 4, 0.51, 0.06, 0.828]]
labels = ['DenseNet', 'Optimiser', 'Learning rate', 'Batchsize', 'Weights', 'Dropout', 'Score']
```

## Setup y-axes

Axes can be linear, logarithmic or categorical. Categorical axes are automatically recognized for string values. Normally only logarithmic axes need to be specified, as in the example below. 

For each axis type you can specify your own labels. It is important that the number of labels matches the number of ticks. For categorical axes the number of ticks is set automatically for the other types they should be specified separately. (For logarithmic axes the number of ticks is always set automatically and cannot be influenced).

The axis limits are determined automatically from the data. If there is not enough data available or if you want to specify your own limits, these must be specified. 

```python
ytype = [[], [], 'log', [], [], [], []]
ylabels = [[], ['Adam', 'Amsgrad', 'RAdam', 'RMSProp', 'SGD+Mom.', 'SGD+Nest.'], 
           [], ['1', '2', '4', '8', '16', '32'], [], [], []]
yticks = [[], [], [], [6], [], [], []]
ylim = [[], [], [1e-06, 0.001], [], [0, 1], [0, 0.25], [0.0, 1.0]]
```

## Plot and Save

Default plot without adjustments:

```python
fig = pcp(data, labels, alpha=0.8)
plt.show()
```

With custom `type`, `ylabels`, `yticks` and `ylim`:

```python
fig = pcp(data, labels, ytype, ylabels, yticks, ylim, alpha=0.8)
plt.savefig('test.svg', format='svg', facecolor="white", bbox_inches='tight')
plt.show()
```
